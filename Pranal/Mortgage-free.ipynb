## Introduction

In this project, I‚Äôm exploring which Pittsburgh neighborhood can be considered the ‚Äúbest‚Äù based on the number of **housing units without a mortgage**.  
The assumption is that neighborhoods with more mortgage-free homeowners reflect higher financial strength, stability, and desirability.  
Owning property outright is a strong economic signal that indicates prosperity and possibly higher property values.

Other metrics could include income, safety, or amenities ‚Äî but I chose this one because it‚Äôs straightforward and data-driven, highlighting neighborhoods with the wealthiest residents.

## Dataset Overview

The dataset used is **`mortgage-status.csv`**, which contains data for housing units in each Pittsburgh neighborhood.  
Key columns include:
- `Neighborhood`: Name of the neighborhood  
- `Estimate; Total:`: Total number of housing units  
- `Estimate; Housing units without a mortgage`: The number of homes owned outright  

Our goal is to identify which neighborhood has the highest number of **mortgage-free homes** and thus ranks as the ‚Äúbest‚Äù under this metric.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("mortgage-status.csv")

# Keep only needed columns
subset = df[['Neighborhood', 'Estimate; Housing units without a mortgage']].copy()

# Clean up column names and data
subset.columns = ['Neighborhood', 'No_Mortgage']
subset['Neighborhood'] = subset['Neighborhood'].fillna("Unknown").str.strip()
subset['No_Mortgage'] = pd.to_numeric(subset['No_Mortgage'], errors='coerce').fillna(0)

# Sort neighborhoods by number of mortgage-free homes
subset = subset.sort_values('No_Mortgage', ascending=False).reset_index(drop=True)

# Display all neighborhoods ranked
print("üèòÔ∏è Mortgage-free homes by neighborhood:")
print(subset.to_string(index=False))

# Plot: Horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(subset['Neighborhood'], subset['No_Mortgage'], color='skyblue')
plt.xlabel("Number of Mortgage-Free Homes")
plt.ylabel("Neighborhood")
plt.title("Mortgage-Free Homes by Neighborhood in Pittsburgh")
plt.gca().invert_yaxis()  # Highest on top
plt.tight_layout()
plt.show()
```

### Reflection: Why Squirrel Hill South is the Best

According to the data, **Squirrel Hill South** ranks as the best neighborhood in Pittsburgh under my metric, the number of housing units owned without a mortgage. 
The high number of mortgage-free homes likely reflects a combination of **generational wealth**, **older homeowners who have paid off their properties**, and **high property desirability** that attracts buyers with the means to purchase without loans.
